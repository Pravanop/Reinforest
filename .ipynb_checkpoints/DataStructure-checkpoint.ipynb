```python
class node:
    def __init__(self,actions,prev): 
        self.actions = actions
        if(prev=='root'):
            self.prev=None
        else : self.prev=prev
        self.next=None
    def printNode(self):
        print("Actions:")
        for action in self.actions:
            action.printAction()
    def setNext(nextNode):
        self.next=nextNode
```

```python
class action:
    def __init__(self, time, action): 
        self.time = time 
        self.action = action
        self.next=None
    def printAction(self):
        print("Action:"+self.action)
        print("Time:"+str(self.time))
```

```python
root=action(800,'Wake up')
one=action(815,'Brush your teeth')
two=action(830,'Workout')
three=action(900,'Eat Breakfast')
four=action(1000,'Work on Project')
five=action(1000,'Work on Irina')
six=action(1000,'Play Witcher')
last=action(1200,'Sleep')
```

```python
root_node=node([root],'root')
node1=node([one],root_node)
node2=node([two],node1)
node3=node([three],node2)
node4=node([four,five],node3)
node5=node([six],node4)
last_node=node([last],node5)
```

```python
#the main data structure: a graph
global_chain = { 
    None:[root_node],
    root_node:[node1],
    node1:[node2,node3],
    node2:[node3],
    node3:[node4,node5],
    node4:[last_node],
    node5:[last_node]
}
```

```python
#printing all nodes
for key,val in global_chain.items():
        for v in val:
            print(v.printNode())
```

```python
# function to generate all possible paths 
def find_all_paths(graph, start, end, path =[]): 
  path = path + [start] 
  if start == end: 
    return [path] 
  paths = [] 
  for node in graph[start]: 
    if node not in path: 
      newpaths = find_all_paths(graph, node, end, path) 
    for newpath in newpaths: 
      paths.append(newpath) 
  return paths 
```

```python
#all possible paths from first node to the last node
for i in (find_all_paths(global_chain,root_node,last_node)):
    print("\nPath:")
    for j in i:
        print(j.printNode())
```
